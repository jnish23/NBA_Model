```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt


from tqdm import tqdm
from nba_api.stats.static import players, teams
from nba_api.stats.library.parameters import SeasonAll
```

```python
seasons = ['20{}-{}'.format(x, x+1) for x in range(10, 21)]

basic_logs = []
for season in seasons:
    log = pd.read_csv('./data/gamelog_{}.csv'.format(season))
    basic_logs.append(log)

all_logs_basic = pd.concat(basic_logs)
all_logs_basic = all_logs_basic[['SEASON_YEAR', 'TEAM_ID', 'TEAM_ABBREVIATION', 'TEAM_NAME', 'GAME_ID',
       'GAME_DATE', 'MATCHUP', 'WL', 'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M',
       'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST',
       'TOV', 'STL', 'BLK', 'BLKA', 'PF', 'PFD', 'PTS', 'PLUS_MINUS']]


advanced_logs = []
for season in seasons:
    log = pd.read_csv('./data/team_advanced_boxscores_{}.csv'.format(season))
    advanced_logs.append(log)

all_logs_advanced = pd.concat(advanced_logs)

scoring_logs = []
for season in seasons:
    log = pd.read_csv('./data/team_scoring_boxscores_{}.csv'.format(season))
    scoring_logs.append(log)
    
all_logs_scoring = pd.concat(scoring_logs)

tracking_logs = []

for season in seasons:
    log = pd.read_csv('./data/team_tracking_boxscores_{}.csv'.format(season))
    tracking_logs.append(log)
    
all_logs_tracking = pd.concat(tracking_logs)

full_df = pd.merge(all_logs_basic, all_logs_advanced, on=['TEAM_ABBREVIATION', 'GAME_ID'], suffixes=[None, '_y'])
full_df = pd.merge(full_df, all_logs_scoring, on=['TEAM_ABBREVIATION', 'GAME_ID'], suffixes=[None, '_y'])
full_df = pd.merge(full_df, all_logs_tracking, on=['TEAM_ABBREVIATION', 'GAME_ID'], suffixes=[None, '_y'])

full_df
```

```python
full_df['WL'] = full_df['WL'].apply(lambda x: 1 if x=='W' else 0)
full_df['record_L10'] = full_df['WL']
full_df['record'] = full_df['WL']
full_df
```

```python
abbr_mapping = {'NJN':'BKN',
               'CHH':'CHA',
               'VAN':'MEM',
               'NOH':'NOP',
               'NOK':'NOP',
               'SEA':'OKC'}

full_df['TEAM_ABBREVIATION'] = full_df['TEAM_ABBREVIATION'].replace(abbr_mapping)
full_df['MATCHUP'] = full_df['MATCHUP'].str.replace('NJN', 'BKN')
full_df['MATCHUP'] = full_df['MATCHUP'].str.replace('CHH', 'CHA')
full_df['MATCHUP'] = full_df['MATCHUP'].str.replace('VAN', 'MEM')
full_df['MATCHUP'] = full_df['MATCHUP'].str.replace('NOH', 'NOP')
full_df['MATCHUP'] = full_df['MATCHUP'].str.replace('NOK', 'NOP')
full_df['MATCHUP'] = full_df['MATCHUP'].str.replace('SEA', 'OKC')


full_df['GAME_DATE'] = pd.to_datetime(full_df['GAME_DATE'])
```

```python
cols_to_drop = []
for col in full_df.columns:
    if col[-2:] == '_y':
        cols_to_drop.append(col)
cols_to_drop
        
full_df = full_df.drop(columns = cols_to_drop)
full_df = full_df.drop(columns = 'TEAM_CITY')
```

```python
def aggregate_team_stats(df = full_df):
    df = full_df.sort_values('GAME_DATE')

    seasons = df['SEASON_YEAR'].unique()


    for season in seasons:
        season_df = df.loc[df['SEASON_YEAR'] == season]
        team_ids = season_df['TEAM_ID'].unique()
        for t_id in tqdm(team_ids, desc='Progress'):
            team_log = season_df.loc[season_df['TEAM_ID'] == t_id]

            avg_team_data = team_log.iloc[:, 8:-1].shift(1).rolling(10, min_periods=1).mean()
            
            record = team_log['record'].shift(1).expanding().mean()

            cols = avg_team_data.columns

            df.loc[(df['SEASON_YEAR'] == season) & (df['TEAM_ID'] == t_id), cols] = avg_team_data
            df.loc[(df['SEASON_YEAR'] == season) & (df['TEAM_ID'] == t_id), 'record'] = record

#     df.to_csv('player_avg_last10.csv', index=False)
        
    return df
    
df = aggregate_team_stats(df=full_df)

df
```
